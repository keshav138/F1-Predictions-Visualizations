```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df_bahrain = pd.read_csv(r'C:\Users\ASUS\Desktop\F1 Predictions & Visualizations\F1-ML-Project\data\clean\season_2021_round_1_clean.csv')
df_bahrain.info()
```

```python
clean = df_bahrain[~(df_bahrain['is_outlap'] | df_bahrain['is_inlap'])].copy()
drivers = ['VER','HAM','BOT','LEC']
df_sel = clean[clean['driver_name'].isin(drivers)].copy()
```

```python
print("AVERAGE SPEED VS LAP TIME")
plt.figure(figsize=(10,7))

sns.lineplot(
    data = df_sel,
    x = 'avg_speed',
    y = 'lap_time',
    hue='driver_name',
    estimator=None,
    lw=2,
    marker='o'
)

plt.title('AVERAGE SPEED VS LAP TIME')
plt.tight_layout()
plt.show()
```

```python
print("AVERAGE THROTTLE VS LAP TIME")
plt.figure(figsize=(10,7))

sns.lineplot(
    data = df_sel,
    x = 'avg_throttle',
    y = 'lap_time',
    hue='driver_name',
    estimator=None,
    lw=2,
    marker='o'
)

plt.title('AVERAGE THROTTLE VS LAP TIME')
plt.tight_layout()
plt.show()
```

```python
print("AVERAGE BRAKE VS LAP TIME")
plt.figure(figsize=(10,7))

sns.lineplot(
    data = df_sel,
    x = 'avg_brake',
    y = 'lap_time',
    hue='driver_name',
    estimator=None,
    lw=2,
    marker='o'
)

plt.title('AVERAGE BRAKE VS LAP TIME')
plt.tight_layout()
plt.show()
```

```python
print('CORRELATION MATRIX')
cols = ['lap_time','avg_speed','avg_throttle','avg_brake','avg_gear','max_rpm']
corr_pd = clean[cols]

plt.figure(figsize=(10,7))
sns.heatmap(corr_pd.corr(),annot=True,cmap='rocket',fmt='.2f',linewidths=0.5)
plt.title('CORRELATION BETWEEN TELEMETRY TERMS')
plt.show()
```
